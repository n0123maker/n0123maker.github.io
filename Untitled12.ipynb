```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
import librosa
import librosa.display

y, sr = librosa.load('D:/Ultra Nate & Roland Clark - The First Time Free (Claptone Remix).mp3')
```

```python
# Compute the spectrogram magnitude and phase
S_full, phase = librosa.magphase(librosa.stft(y))

idx = slice(*librosa.time_to_frames([0, 5], sr=sr))
plt.figure(figsize=(16, 4))
librosa.display.specshow(librosa.amplitude_to_db(S_full[:, idx], ref=np.max), y_axis='log', x_axis='time', sr=sr)
plt.colorbar()
plt.tight_layout()
```
